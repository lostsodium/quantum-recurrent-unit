```python
import pickle
import jax.numpy as jnp
import csv
```

```python
nb_name = 'I1_H8_O2 (J8 3 layers rx_ry 100x LOOCV) seed_0'
```

```python
n = 569
output_csv = nb_name + '_wdbc_loss_history.csv'

# ── 收集所有 sample 的 loss ──────────────────────────────
all_loss = {}

for i in range(n):
    load_name = nb_name + '_' + f"{i:03}"
    try:
        with open(load_name + '_current.pkl', 'rb') as f:
            (locLossList, locLoss, vlocLoss, lossList, vlossList, loss,
             minLoss, vminLoss, vlocAcc, vmaxAcc, params, loc_params, vloc_params, best_params,
             vbest_params, loc_params_list, vloc_params_list, reset_steps,
             acc_results, _i, key2, train_dataset, valid_dataset, test_dataset,
             opt_state, update_lr, learning_rate, all_loss_list,
             all_vloss_list, dataset_indices, ES_checker) = pickle.load(f)
        all_loss[f'sample{i:03d}'] = [float(x) for x in lossList]
    except:
        print(f'Error: {load_name}')

print(f'Loaded: {len(all_loss)} / {n} samples')

# ── 寫入 CSV ──────────────────────────────────────────────
col_names  = list(all_loss.keys())
max_epochs = max(len(v) for v in all_loss.values())

with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['epoch'] + col_names)
    for ep in range(max_epochs):
        row = [ep + 1]
        for k in col_names:
            row.append(all_loss[k][ep] if ep < len(all_loss[k]) else '')
        writer.writerow(row)

print(f'Saved: {output_csv}')
print(f'Shape: {max_epochs} epochs × {len(all_loss)} samples')
```
