```python
import pickle
import jax.numpy as jnp
```

```python
nb_name = 'I1_H8_O2 (J8 3 layers rx_ry 100x LOOCV) seed_0'
```

```python
loocv_results = []
sum_acc = jnp.zeros(4)
n = 569
m = 0
for i in range(n):
    load_name = nb_name + '_' + f"{i:03}"
    try:
        with open(load_name+'_current.pkl', 'rb') as f:
            (locLossList, locLoss, vlocLoss, lossList, vlossList, loss,
             minLoss, vminLoss, vlocAcc, vmaxAcc, params, loc_params, vloc_params, best_params,
             vbest_params, loc_params_list, vloc_params_list, reset_steps,
             acc_results, i, key2, train_dataset, valid_dataset, test_dataset,
             opt_state, update_lr, learning_rate, all_loss_list, 
             all_vloss_list, dataset_indices, ES_checker) = pickle.load(f)
            sum_acc += jnp.array(acc_results[0])
            m += 1
            loocv_results.append(acc_results[0] + [len(lossList)] + [test_dataset[0][1]])
    except:
        print('Error: ' + load_name)
        
loocv_results = jnp.array(loocv_results)
sum_acc / m
```

```python
import csv
```

```python
with open(nb_name+'_results.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Train acc. (trainging best)', 'Test acc. (trainging best)', 
                     'Train acc. (validation best)', 'Test acc. (validation best)',
                    'Epochs (x10)', 'True label'])
    for _ in loocv_results:
        writer.writerow(_)
```
