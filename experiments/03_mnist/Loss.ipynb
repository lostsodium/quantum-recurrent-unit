```python
import jax.numpy as jnp
import pickle
import os
import csv
```

```python
nb_name = 'MNIST 3_5 (J8 I8_H8_L4_O8) SEED_2'
```

```python
# ---- 收集所有 fold 的 loss ----
all_train = {}

for a in range(7):
    for b in range(1):
        load_name = nb_name + '_' + str(a+1) + '-' + str(b+1)
        print(load_name)
        with open(load_name+'_current.pkl', 'rb') as f:
            (locLossList, locLoss, vlocLoss, lossList, vlossList, loss,
             minLoss, vminLoss, vlocAcc, vmaxAcc, params, loc_params, vloc_params, best_params,
             vbest_params, loc_params_list, vloc_params_list, reset_steps,
             acc_results, i, key2, train_dataset, valid_dataset, test_dataset,
             opt_state, update_lr, learning_rate, all_loss_list, 
             all_vloss_list, dataset_indices, ES_checker, avg_vacc, avg_vloss, avg_v_i) = pickle.load(f)
        key = f'fold{a+1}_run{b+1}'
        all_train[key] = [float(x) for x in lossList]
        print(f'  Epochs: {len(lossList)}')
#         fold_results.append(acc_results[0])
# fold_results = jnp.array(fold_results)
# fold_results
```

```python
# ---- 找最大 epoch 數 ----
max_epochs = max(len(v) for v in all_train.values())

# ---- 寫入 CSV ----
fold_keys = list(all_train.keys())
output_csv = nb_name + '_loss_history.csv'

with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)

    header = ['epoch'] + fold_keys
    writer.writerow(header)

    for ep in range(max_epochs):
        row = [ep + 1]
        for k in fold_keys:
            row.append(all_train[k][ep] if ep < len(all_train[k]) else '')
        writer.writerow(row)

print(f'\nSaved to: {output_csv}')
```
